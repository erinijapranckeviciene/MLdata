Synthetic data blobs - Multivariate normal  

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
# standard deviation controls spread of points around the centroids 
c_std=[1,1]
X, y = make_blobs(n_samples=n_samples, random_state=random_state, centers=2,n_features=2,cluster_std=c_std)
```

```
# y contains class labels of the blobs
print(y)
```

```
plt.figure(figsize=(12, 12))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Synthetic Blobs")
plt.legend()
plt.grid(True)
plt.show()
```
